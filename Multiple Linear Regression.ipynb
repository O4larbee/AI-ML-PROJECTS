## 🔍 **Regression Assumptions & Interpretation**

### Assumption Checks:

* **Durbin-Watson (1.873)** suggests no major autocorrelation in residuals.
* **Jarque-Bera and Skew/Kurtosis** values indicate reasonable normality of residuals.
* Model diagnostics (e.g., Cond. No. = 144) suggest **no severe multicollinearity**.

### Interpreting Coefficients:

* **TV (Large)** and **TV (Medium)** have **strong negative coefficients**, indicating diminishing or negative returns at high ad spend levels.
* **Influencer (Mega)** and **Micro influencers** show **insignificant p-values**, meaning their contribution to sales isn't statistically strong.
* **Radio** has a **strong, significant positive coefficient** (*2.95*, p < 0.001) — a key contributor to sales.
* **Intercept (217.37)** indicates base sales when all channels are at 0.
